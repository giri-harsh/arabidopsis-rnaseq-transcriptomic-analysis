# 05 — Biological Interpretation

Inspect the top significant DEGs and frame what further biological annotation (outside this pipeline's scope) would be needed.

```python
import sys
sys.path.insert(0, '../src')

from rnaseq.io import load_dataset
from rnaseq.config import PipelineConfig
from rnaseq.preprocessing import run_preprocessing
from rnaseq.deseq_pipeline import run_deseq2
from rnaseq.statistics import extract_significant_degs

dataset = load_dataset('../data/example_dataset/counts.csv', '../data/example_dataset/metadata.csv', condition_column='condition')
config = PipelineConfig(counts_path='x', metadata_path='y', reference_level='control')
preprocessing = run_preprocessing(dataset.counts, config)
deseq_result = run_deseq2(preprocessing.filtered_counts, dataset.metadata, config)
deg_summary = extract_significant_degs(deseq_result.results_df, config)
```

```python
top_up = deg_summary.upregulated.sort_values('padj').head(10)
top_down = deg_summary.downregulated.sort_values('padj').head(10)
print('Top upregulated:'); display(top_up)
print('Top downregulated:'); display(top_down)
```

## Interpretation notes

- Genes with large positive log2 fold change are candidates for stress-induced upregulation; large negative fold-change genes are candidates for stress-induced repression.
- This notebook does **not** perform gene ontology or pathway enrichment — gene IDs above would need to be cross-referenced against TAIR / Ensembl Plants (for real Arabidopsis gene identifiers) to attach biological function.
- The gene IDs shown here (`AT_SIM_*`) come from the bundled **synthetic** example dataset and have no real biological meaning; this notebook demonstrates the interpretation workflow structure, not an actual biological finding.
