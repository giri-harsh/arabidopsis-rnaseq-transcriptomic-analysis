# 01 — Data Loading

Load the raw count matrix and sample metadata, and run the validation checks in `rnaseq.io` before any analysis proceeds.

```python
import sys
sys.path.insert(0, '../src')

from rnaseq.io import load_dataset

COUNTS_PATH = '../data/example_dataset/counts.csv'
METADATA_PATH = '../data/example_dataset/metadata.csv'
```

```python
dataset = load_dataset(COUNTS_PATH, METADATA_PATH, condition_column='condition')
print(f'{dataset.n_genes} genes x {dataset.n_samples} samples')
dataset.counts.head()
```

`load_dataset` aligns counts and metadata, checks for duplicate genes, negative/missing values, and confirms at least two condition levels are present — see `rnaseq/io.py` for the full validation contract.
